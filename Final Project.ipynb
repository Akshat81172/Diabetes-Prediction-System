```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = {
    'gender': np.random.choice(['Male', 'Female'], size=100),
    'age': np.random.randint(20, 80, size=100),
    'hypertension': np.random.choice([0, 1], size=100),
    'heart_disease': np.random.choice([0, 1], size=100),
    'smoking_history': np.random.choice(['never', 'former', 'current'], size=100),
    'bmi': np.random.uniform(18, 35, size=100),
    'HbA1c_level': np.random.uniform(4, 10, size=100),
    'blood_glucose_level': np.random.uniform(70, 200, size=100),
    'diabetes': np.random.choice([0, 1], size=100)
}

df = pd.DataFrame(data)

# Convert categorical data to numerical values for simplicity
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
df['smoking_history'] = df['smoking_history'].map({'never': 0, 'former': 1, 'current': 2})

# Calculate mean values for diabetic and non-diabetic individuals
means = df.groupby('diabetes').mean()

# Plotting
features = ['gender', 'age', 'hypertension', 'heart_disease', 'smoking_history', 'bmi', 'HbA1c_level', 'blood_glucose_level']
diabetes_status = ['Non-Diabetic', 'Diabetic']

x = np.arange(len(features))  # Label locations
width = 0.35  # Bar width

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, means.loc[0, features], width, label='Non-Diabetic')
bars2 = ax.bar(x + width/2, means.loc[1, features], width, label='Diabetic')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Features')
ax.set_ylabel('Mean Values')
ax.set_title('Comparison of Feature Means for Diabetic and Non-Diabetic Individuals')
ax.set_xticks(x)
ax.set_xticklabels(features)
ax.legend()

# Add value labels on top of the bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)

fig.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
